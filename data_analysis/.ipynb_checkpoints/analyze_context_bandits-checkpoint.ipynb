```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import logit
from glob import glob
plt.rcParams["figure.figsize"] = (15, 10)
plt.rcParams.update({'font.size': 22})
```

```python
def convert_response(row):
    if row['response'] == 49:
        return 1
    elif row['response'] == 50:
        return 2
    elif row['response'] == 51:
        return 3
    else:
        return np.nan
    
def clean_data(file):
    df = pd.read_csv(file)
    clean_df = df.loc[:,['subject_id','trial_type','response','trial_n','context_n','outcome','is_valid','probed_trial', 'probed_context', 'correct']]
    clean_df['choice'] = clean_df.apply (lambda row: convert_response(row), axis=1)
    
    return clean_df 

def to_probability(odds):
    return odds/(1+odds)

def prob_choice(param):
    return to_probability(np.exp(param))
    
# make context value dict
def make_context_val(data):
    context_val = {}
    for context in range(6):
        curr_context_val_dict = {}
        curr_context = data.query("context_n=="+str(context))
        for choice in range(1,4):
            choice_total = len(curr_context.query("choice == "+ str(choice)))
            choice_reward = len(curr_context.query("choice == "+str(choice)+" & outcome == 1"))
            choice_not_reward = len(curr_context.query("choice == "+str(choice)+" & outcome == 0"))
            if choice_total > 0:
                value = (choice_reward - choice_not_reward)/choice_total
            else:
                value = 0
            curr_context_val_dict[choice] = value
        context_val[context] = curr_context_val_dict
    return context_val
```

```python
def perseveration(test_data, curr_trial_data, curr_trial_num, n_trial_back, choice_focus):
    # did you make the this choice n trials ago? 
    # test_data: all the data from previous test phase trials (only choice trials)
    # curr_trial_data: data from a single choice trial from the test phase 
    # curr_trial_num: the choice trial number
    # trial_back: how many trials back back do we want to compare the current choice to
    # choice focus: the choice we're comparing this all to 
    
    if (curr_trial_num - n_trial_back) < 0:
        return False
    else:
        prev_trial_choice = test_data.loc[curr_trial_num - n_trial_back].choice
        return choice_focus == prev_trial_choice

def n_prev_reward(test_data, curr_trial_data,curr_trial_num,n_trial_back, choice_focus):
    # were you rewarded for making this choice 
    if (curr_trial_num - n_trial_back) < 0:
        return False
    else:
        prev_reward = test_data.loc[curr_trial_num-n_trial_back].outcome 
        return bool(prev_reward)  & perseveration(test_data,curr_trial_data,curr_trial_num, n_trial_back, choice_focus)

def probe_check(all_data,curr_trial_data):
    whole_exp_trial_n = curr_trial_data.trial_n
    prev_trial = all_data.query("trial_n==(@whole_exp_trial_n-1)").reset_index()
    after_valid_probe = (prev_trial.at[0,"is_valid"] == 1.0) 
    return after_valid_probe


def probe_influence(all_data, test_data, curr_trial_data, context_val_dict, choice_focus):
    # check if this trial was after a recog trial
    if curr_trial_data.trial_n == 240:
        return 0, 0, 0
    else:  
        if probe_check(all_data,curr_trial_data): # checks valid probe 
            recog_trial = all_data.query("trial_n == (@curr_trial_data.trial_n-1)").reset_index()
            probed_trial_n = recog_trial.at[0,"probed_trial"]
            probed_trial = all_data.query("trial_n == @probed_trial_n").reset_index()
            
            probed_trial_choice = probed_trial.at[0,"choice"]
            probe_choice = probed_trial_choice == choice_focus
            probe_outcome = probe_choice & bool(probed_trial.at[0,"outcome"])
            probe_context = probed_trial.at[0,"context_n"]
            probe_context_val = context_val_dict[probe_context][choice_focus]
            
            return probe_choice, probe_outcome, probe_context_val 
        else:
            return 0, 0, 0 
```

```python
def make_feature_matrix(filename,all_data,test_phase_data):
    context_val_dict = make_context_val(all_data)

    reg_arr = np.zeros((7,3,120))
    choices = np.zeros((120,1))
    for option in range(1,4):
        for index, row in test_phase_data.iterrows():
            choices[index] = row.choice # DV
        
            reg_arr[0,option-1,index] = perseveration(test_phase_data, row, index, 1, option) # IV 1
        
            for trial_back in range(1,4): # IV 2-4
                reg_arr[trial_back,option-1,index] = n_prev_reward(test_phase_data,row,index,trial_back, option)  
            probe_choice, probe_outcome, probe_context_val = probe_influence(all_data, test_phase_data, 
                                                                        row, context_val_dict, option)
            reg_arr[4,option-1,index] = probe_choice
            reg_arr[5,option-1,index] = probe_outcome
            reg_arr[6,option-1,index] = probe_context_val

    # drop the no choice trials 
    no_choice_trials = np.where(np.isnan(choices))
    no_choice_trials = no_choice_trials[0]
    
    # trials wherre the probe was not recalled
    low_conf = [53,54,55]
    missed_trials = all_data.query("trial_type=='do_recog' & response in @low_conf & is_valid == 1")
    missed_trials = missed_trials.trial_n
    trial_after = [x+1 for x in missed_trials if x < 419] 

    test_phase_index = []

    for t in trial_after:
        ind = test_phase_data.query("trial_n== @t").reset_index().index[0]
        test_phase_index.append(ind)
    
    # add all dropped trials together
    to_drop = list(no_choice_trials) + test_phase_index
    to_drop = np.unique(to_drop)

    reg_arr=np.delete(reg_arr,to_drop,axis=2)
    choices=np.delete(choices,to_drop,axis=0)
    

    # now zscore for each predictor 
    for pred in range(7):
        reg_arr[pred,:,:] = stats.zscore(reg_arr[pred,:,:],axis=None)

    
    np.savez(filename,reg_arr,choices)
    return 
```

```python
def sub_best_pirate(data):
    pick_trials=data.query("trial_type=='do_pick_best_pirate'")
    context_val_dict = make_context_val(data)

    score = []
    for context in range(6):
        their_pick = pick_trials.iloc[context].choice
        true_best = max(context_val_dict[context], key=context_val_dict[context].get)
        score.append(their_pick==true_best)
    
    return np.mean(score)
```

```python
filenames = glob('../data/*csv')

pick_pirate = pd.DataFrame(columns=['sub_num','score'])
for i,f in enumerate(filenames):
    data = clean_data(f)
    sub_id = data.subject_id.dropna().reset_index().subject_id[0]
    score = sub_best_pirate(data)
    temp = pd.DataFrame({'sub_num':"sub_"+str(i),'score':score},index=[0])
    pick_pirate = pd.concat([pick_pirate,temp])
    
    
    
    
```

```python
def calc_dprime(data):
    z_trans = stats.norm.ppf
    
    recog_trials = data.query("trial_type=='do_recog'")
    hits = len(recog_trials.query('is_valid==1 & correct==1'))
    misses = len(recog_trials.query('is_valid==1 & correct==0'))
    hit_rate = hits/(hits+misses)
    
    crs = len(recog_trials.query('is_valid==0 & correct==1'))
    fas =len(recog_trials.query('is_valid==0 & correct==0'))
    fa_rate = fas/(fas+crs)
    
    half_hit = 0.5 / (hits + misses)
    half_fa = 0.5 / (fas + crs)
    
    if hit_rate == 1: 
        hit_rate = 1 - half_hit
    if hit_rate == 0: 
        hit_rate = half_hit
        
    if fa_rate == 1: 
        fa_rate = 1 - half_fa
    if fa_rate == 0: 
        fa_rate = half_fa
    
    return z_trans(hit_rate) - z_trans(fa_rate)
    
    
```

```python
recog = pd.DataFrame(columns=['sub_num','d_prime'])
for i,f in enumerate(filenames):
    data = clean_data(f)
    d_prime = calc_dprime(data)
    temp = pd.DataFrame({'sub_num':"sub_"+str(i),'d_prime':d_prime},index=[0])
    recog = pd.concat([recog,temp])
```
