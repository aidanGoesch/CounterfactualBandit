```julia
using Distributions: Categorical
using NPZ
using Glob

include("fit_softmax.jl")
```

```julia
for f in files:
    
    # read in file 
    y = npzread(f);
    
    # extract relevant variables 
    choices = y["arr_1"];
    choices = convert(Array{Int64}, choices);
    mat = y["arr_0"];
    
    # run regression 
    @time β_fit = fit_softmax_model(mat, choices)
    
    # save file
    output_folder = "sub_reg_coef/"
    filename = split(f,"/")[2]
    output_path = output_folder + filename 
    
    npzwrite(output_path, β_fit)
    
    
```
